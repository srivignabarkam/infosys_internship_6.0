```python
#Fill text columns with "unknown"
text_columns=df.select_dtypes(include=['object']).columns
for col in text_columns:
  if df[col].isnull().sum()>0:
    df[col]=df[col].fillna('Unknown')
    print(f"Filled '{col}' with 'Unknown'")


#Fill numeric columns with median
numeric_columns=df.select_dtypes(include=[np.number]).columns
for col in numeric_columns:
  if df[col].isnull().sum()>0:
    median_val=df[col].median()
    df[col]=df[col].fillna(median_val)
    print(f" Filed '{col}' with median: {median_val} ")
df.head()
```

Remove Duplicates

```python
#Remove Dupicates
print("Remove Duplicate")
print(f"Number of rows before removing duplicates: {len(df)}")
duplicate_count=df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

df=df.drop_duplicates()
print(f"Number of rows after removing duplicates: {len(df)}")
```

Standardize column names

```python
print("Standardizing column names:")
print()

df.columns=df.columns.str.lower().str.replace(' ','_')
print(f"Standardized names: \n{list(df.columns)}")
print()
df
```

Remove Invalid Values

```python
print("Removing Invalid Values:")
print()
if 'customer_age' in df.columns:
  initial=len(df)
  df=df[df['customer_age']>=18]
  print(f"Removed {initial-len(df)} rows with age<18")

df=df.dropna(how='all')
print(f"Removed Completely empty rows")
```

```python
#Data cleaning
print(f"Final size:{len(df)} rows x {len(df.columns)} columns")
print()

df.to_csv('01_cleaned_data.csv',index=False)
```

# DATA TRANSFORMATION

```python
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
from datetime import datetime

df=pd.read_csv('02_validated_data.csv')

print(df.dtypes)
print()

print("Converting data types:")

if 'ticket_status' in df.columns:
  df['ticket_status']=df['ticket_status'].astype('category')
  print(f"Converted ticket_status to category")

if 'customer_age' in df.columns:
  df['customer_age']=df['customer_age'].astype('int32')
  print(f"Converted customer_age to int32")

print()
print(df.dtypes)
```

```python
print("Parsing Data columns:")
print()

date_columns=['date_of_purchase','first_response_time',]
for col in date_columns:
  if col in df.columns and df[col].dtype=='object':
    df[col]=pd.to_datetime(df[col],errors='coerce')
    print(f"converted '{col}' to datetime")
```

Creating new features

```python
print("Creating New Features")
print()

if 'date_of_purchase' in df.columns:
  df['purchase_year']=df['date_of_purchase'].dt.year
  df['purchase_month']=df['date_of_purchase'].dt.month
  df['purchase_day']=df['date_of_purchase'].dt.day
df
```

Encoding Categorical Data


```python
from pandas.core.arrays import categorical
print("Encoding Categorical Data:")
print()

categorical_col=df.select_dtypes(include=['object']).columns
encoders={}

for col in categorical_col:
  encoder=LabelEncoder()
  df[col + '_encoded']=encoder.fit_transform(df[col].astype(str))
  encoders[col]=encoder
  print(f"Encoded '{col}'")
```

Normalization

```python
print("Normalization")
print()

if 'customer_age' in df.columns:
  scaler=MinMaxScaler()
  df['customer_age_normalized']=scaler.fit_transform(df[['customer_age']])
  print(f"Normalized 'customer_age'")
  print(f"Range: {df['customer_age_normalized'].min()},{df['customer_age_normalized'].max()}")
```

```python
print("Transformation Data")
print(f"Final Size: {len(df)} rows x {len(df.columns)} columns")
print()

df.to_csv('03_transformed_data.csv',index=False)
```

# DATA FILTERING

Remove unnecessary columns

```python
df=pd.read_csv('03_transformed_data.csv')

print("Removing Unnecessary Columns:")
print()

columns_to_remove=['customer_name','customer_email']
for col in columns_to_remove:
  if col in df.columns:
    df=df.drop(columns=[col])
    print(f"Removed '{col}'")

text_cols=df.select_dtypes(include=['object']).columns
for col in text_cols:
  if col + '_encoded' in df.columns:
    df=df.drop(columns=[col])
    print(f"Removed: '{col}")

df
```

Filter Records by Condition

```python
print("Filtering Records by Condition:")
print()

if 'customer_satisfaction_rating' in df.columns:
  df=df.dropna(subset=['customer_satisfaction_rating'])
  print(f"Removed rows with missing 'customer_satisfaction_rating'")

if 'purchase_year' in df.columns:
  recent_year=df['purchase_year'].max()-2
  df=df[df['purchase_year']>=recent_year]
  print(f"Removed records from before {recent_year}")
df
```

Remove Incomplete Records

```python
print("Removing Incomplete Records:")
print()

intial=len(df)
critical_fields=['customer_age','ticket_status','ticket_priority']
df=df.dropna(subset=[col for col in critical_fields if col in df.columns])
print(f"Removed {intial-len(df)} rows with missing critical fields")

print(f"Final Size: {len(df)} rows x {len(df.columns)} columns")
```

```python
print("Filtered Data")
print()
df.to_csv('04_filtered_data.csv',index=False)
```

# DATA ENRICHMENT

Create Derived Fields

```python
df=pd.read_csv('04_filtered_data.csv')

print()

def categorize_age(age):
  if age<30:
    return 'Young'
  elif age<50:
    return 'Middle-aged'
  else:
    return 'Senior'

if 'customer_age' in df.columns:
  df['age_group']=df['customer_age'].apply(categorize_age)
  print("Created 'age_group' derived field")
  print(f"Categories: {df['age_group'].unique()}")
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")
plt.figure(figsize=(10,6))

age_order=['Young','Middle-aged','senior']

ax=sns.countplot(data=df,x='age_group',order=age_order,palette='viridis')

plt.title('Customer Age Distribution')
plt.xlabel('Age Group')
plt.ylabel('Number of Customers')

for p in ax.patches:
  ax.annotate(f'{int(p.get_height())}',
              (p.get_x()+p.get_width()/2.,p.get_height()),
              ha='center',va='center',
              xytext=(0,9),
              textcoords='offset points')
plt.show()
```

```python
#Lookup based enrichment
priority_mapping={
    'low':1,
    'medium':2,
    'high':3,
    'critical':4
}
print("created")
```

```python

#interaction Features
print("Creating interaction features")
print()

if 'customer_age' in df.columns and 'customer_satisfaction_rating' in df.columns:
  df['customer_value_score']=(df['customer_age']*df['customer_satisfaction_rating'])
  print("Created 'customer_value_score' interaction feature")
```

Time based Features

```python
print("Extracting time features")
print()

if 'purchase_month' in df.columns:
  def get_season(month):
    if month in [12,1,2]:
      return 'Winter'
    elif month in [3,4,5]:
      return 'Spring'
    elif month in [6,7,8]:
      return 'Summer'
    else:
      return 'Fall'
df['season']=df['purchase_month'].apply(get_season)
print("Created 'season' time-based feature")
print(f"Categories: {df['season'].unique()}")
```

```python
print("Data Enrichment")
print(f"Final Size: {len(df)} rows x {len(df.columns)} columns")
print()

df.to_csv('05_enrichede_data.csv',index=False)
```
